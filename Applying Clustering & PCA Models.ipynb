```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
%matplotlib inline
```

```python
dataframe_bank = pd.read_csv("bank3.csv")
dataframe_bank
```

```python
dataframe_bank.drop('deposit', inplace = True, axis = 1)
dataframe_bank
```

```python
#Changing the data type of categorical data.
dataframe_bank["campaign"] = dataframe_bank["campaign"].astype('category')
dataframe_bank["job"] = dataframe_bank["job"].astype("category")
dataframe_bank["marital"] = dataframe_bank["marital"].astype("category")
dataframe_bank["education"] = dataframe_bank["education"].astype("category")
dataframe_bank["housing"] = dataframe_bank["housing"].astype("category")
dataframe_bank["loan"] = dataframe_bank["loan"].astype("category")
dataframe_bank["poutcome"] = dataframe_bank["poutcome"].astype("category")
dataframe_bank
```

```python
#Changing all categorical values to numerical data.
dataframe_bank["campaign"] = dataframe_bank["campaign"].cat.codes
dataframe_bank["job"] = dataframe_bank["job"].cat.codes
dataframe_bank["marital"] = dataframe_bank["marital"].cat.codes
dataframe_bank["education"] = dataframe_bank["education"].cat.codes
dataframe_bank["housing"] = dataframe_bank["housing"].cat.codes
dataframe_bank["loan"] = dataframe_bank["loan"].cat.codes
dataframe_bank["poutcome"] = dataframe_bank["poutcome"].cat.codes
dataframe_bank
```

## KMEANS Clustering

```python
kmeans_model = KMeans(n_clusters=5).fit(dataframe_bank)

centroids = kmeans_model.cluster_centers_
labels = kmeans_model.labels_

print(centroids)
print(labels)
```

```python
errors = []
for k in range(2, 9):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(dataframe_bank)
    errors.append(kmeans.inertia_)
    
    
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 9), errors)
plt.grid(True)
```

```python
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
visualizer = KElbowVisualizer(kmeans_model, k=(2,9))

visualizer.fit(dataframe_bank)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
#The optimum k value for this set of clusters and for this dataset is k= 4. 
#Here the n number of clusters during the model training was 5 but the k mean ranged between 2 and 9. 
```

## Hierarchial Clustering

```python
from sklearn.preprocessing import normalize
scaled = normalize(dataframe_bank)
scaled = pd.DataFrame(scaled, columns=dataframe_bank.columns)
scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 8))  
plt.title("Dendrogram")  
dendogram = shc.dendrogram(shc.linkage(scaled, method='complete'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')  
clusters.fit_predict(scaled)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(scaled['age'], scaled['education'], c=clusters.labels_) 
```
